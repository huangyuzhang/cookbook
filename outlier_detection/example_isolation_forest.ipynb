# Outlier Detection with Isolation Forest

*Explanation comments added.

```python
# importing libaries ----
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import savefig
from sklearn.ensemble import IsolationForest

# plot graphs in retina resolution
%config InlineBackend.figure_format = 'retina'

# ignore future warnings from sklearn and seaborn
import warnings
warnings.simplefilter(action = 'ignore', category=FutureWarning)
warnings.filterwarnings('ignore')
def ignore_warn(*args, **kwargs):
    pass

warnings.warn = ignore_warn

# default plot settings
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
# Generating data ----

rng = np.random.RandomState(42)

# Generating training data 

# For random samples from N(\mu, \sigma^2), use:
# sigma * np.random.randn(...) + mu
# ref: https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.random.randn.html

X_train = 0.2 * rng.randn(1000, 2) # mu = 0, sigma = 0.2
X_train = np.r_[X_train + 3, X_train] # row concatenation, generate 2 clusters
X_train = pd.DataFrame(X_train, columns = ['x1', 'x2']) # form df with column names

# Generating new, 'normal' observation
X_test = 0.2 * rng.randn(200, 2)
X_test = np.r_[X_test + 3, X_test]
X_test = pd.DataFrame(X_test, columns = ['x1', 'x2'])

# Generating outliers (range: -1 ~ 5)
X_outliers = rng.uniform(low=-1, high=5, size=(50, 2))
X_outliers = pd.DataFrame(X_outliers, columns = ['x1', 'x2'])
```

```python
# Plotting generated data ----

plt.title("Data")

p1 = plt.scatter(X_train.x1, X_train.x2, c='white',
                 s=20*4, edgecolor='k', alpha=0.5)
p2 = plt.scatter(X_test.x1, X_test.x2, c='green',
                 s=20*4, edgecolor='k', alpha=0.5)
p3 = plt.scatter(X_outliers.x1, X_outliers.x2, c='red',
                s=20*4, edgecolor='k', alpha=0.5)

plt.axis('tight')
plt.xlim((-2, 5))
plt.ylim((-2, 5))
plt.legend([p1, p2, p3],
           ["training observations",
            "new regular obs.", "new abnormal obs."],
           loc="lower right")

# saving the figure
# plt.savefig('generated_data.png', dpi=300)

plt.show()
```

```python
# Isolation Forest ----

# training the model
clf = IsolationForest(max_samples=100, contamination = 0.1, random_state=rng)
# the contamination parameter specifies the percentage of observations we believe to be outliers
# max_samples defines the maximum depth

clf.fit(X_train)

# predictions
# doc: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.IsolationForest.html#sklearn.ensemble.IsolationForest.predict
y_pred_train = clf.predict(X_train)
y_pred_test = clf.predict(X_test) # 1 = inlier; -1 = outlier
y_pred_outliers = clf.predict(X_outliers)
```

```python
# new, 'normal' observations ("1")
print("Accuracy:", list(y_pred_test).count(1)/y_pred_test.shape[0])
```

```python
# outliers("-1")
print("Accuracy:", list(y_pred_outliers).count(-1)/y_pred_outliers.shape[0])
```

```python
# Inspecting the outliers ----

# adding the predicted label
X_outliers = X_outliers.assign(y = y_pred_outliers)

plt.title("Outlier Inspection")

p1 = plt.scatter(X_train.x1, X_train.x2, c='white',
                 s=20*4, edgecolor='k', alpha=0.5)
p2 = plt.scatter(X_outliers.loc[X_outliers.y == -1, ['x1']], 
                 X_outliers.loc[X_outliers.y == -1, ['x2']], 
                 c='red', s=20*4, edgecolor='k', alpha=0.5)
p3 = plt.scatter(X_outliers.loc[X_outliers.y == 1, ['x1']], 
                 X_outliers.loc[X_outliers.y == 1, ['x2']], 
                 c='green', s=20*4, edgecolor='k', alpha=0.5)

plt.axis('tight')
plt.xlim((-2, 5))
plt.ylim((-2, 5))
plt.legend([p1, p2, p3],
           ["training observations",
            "detected outliers", 
            "detected regular obs."],
           loc="lower right")

# saving the figure
# plt.savefig('outlier_inspection.png', dpi=300)

plt.show()
```
